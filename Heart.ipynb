```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df2 = df.groupby(["Gender","Heart Disease"]).size().unstack()

df2.plot(kind="bar", stacked=True, figsize=(10, 5), colormap="viridis")

plt.xlabel("Gender")
plt.ylabel("Count")

plt.title("Heart Disease Cases by Gender")

plt.legend(["No Heart Disease", "Heart Disease"])

plt.xticks(rotation=0)
plt.show()

```

```python
grouped_data = df.groupby("Age")["Heart Disease"].max().tail(30)
plt.figure(figsize=(12, 6))
grouped_data.plot(kind="bar", colormap="viridis")

plt.xlabel("Age")
plt.ylabel("Number of Heart Disease Cases")
plt.title("Age vs Heart Disease Cases")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

```python
exercise_group = df.groupby("Exercise Hours")["Heart Disease"].mean()

plt.figure(figsize=(8, 5))

plt.scatter(exercise_group.index, exercise_group.values, color="red", alpha=0.7)

plt.xlabel("Exercise Hours per Week")

plt.ylabel("Average Heart Disease Cases")
plt.title("Heart Disease vs. Exercise Hours")
plt.show()
```

```python
age_heart_disease = df.groupby("Age")["Heart Disease"].sum()

plt.figure(figsize=(8, 10))
age_heart_disease.plot(kind="pie",autopct='%1.1f%%', cmap="coolwarm")


plt.title("Distribution of Heart Disease by Age")

plt.show()
```

```python
columns = ["Age", "Cholesterol", "Blood Pressure", "Heart Rate", "Exercise Hours", "Stress Level", "Heart Disease"]

df_selected = df[columns]

df_grouped = df_selected.groupby("Age").max()

corr_matrix = df_grouped.corr()

plt.figure(figsize=(10, 6))

sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)


plt.title("Correlation Heatmap of Selected Features by Age")

plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder 

lable_encoder =LabelEncoder()

df["Gender_na"] = lable_encoder.fit_transform(df["Gender"])
df["Family History_na"] = lable_encoder.fit_transform(df["Family History"])
df["Diabetes_na"] = lable_encoder.fit_transform(df["Diabetes"])
df["Obesity_na"] = lable_encoder.fit_transform(df["Obesity"])
```

```python
df.head()
```

```python
df2=["Age","Gender","Gender_na","Cholesterol","Blood Pressure","Heart Rate","Smoking","Exercise Hours","Stress Level","Heart Disease",]
```

```python
df=df[df2]
```

```python
df.head()
```

```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression

features = ['Age', 'Gender_na', 'Cholesterol', 'Blood Pressure', 'Heart Rate', 'Exercise Hours', 'Stress Level']
X = df[features].values  
y = df["Heart Disease"] 


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)


model = LinearRegression()
model.fit(X_train, y_train)

new_patient = [[69,0,192,90,72,4,7]] 
prediction = model.predict(new_patient)

```

```python
if prediction[0]== 1:
    print("Predicted Heart Disease: Yes")
else:
    print("Predicted Heart Disease: No")
```
